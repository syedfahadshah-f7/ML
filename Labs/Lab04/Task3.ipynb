```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r"D:\Sem 5\ML LABS\DataSets\cancer_patient.csv")
display(df)
```

```python
print(df['Level'].value_counts())
```

```python
print(df.duplicated().sum())
```

```python
df.drop(columns=['Patient Id','index'],inplace=True)
le = LabelEncoder()
Y = df.pop('Level')
Y = le.fit_transform(Y)
X = df
display(X)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.8,random_state=0,stratify=Y)

X_train_val,X_test_val,Y_train_val,Y_test_val = train_test_split(X_train,Y_train,train_size=0.7,random_state=0,stratify=Y_train)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train_val,Y_train_val)
pred_val = dt.predict(X_train_val)
tr_acc = accuracy_score(Y_train_val,pred_val)
pred = dt.predict(X_test)
te_acc = accuracy_score(Y_test,pred)
print(f"Traning Accuracy: {tr_acc}, Testing Accuracy: {te_acc}")
```
