```python
# Task 03

# Correlation Graph
df5 = df.copy(deep=True)
df5.drop(['Patient Id','Level'],axis=1,inplace=True)
plt.figure(figsize=(12,7))
snb.heatmap(data=df5.corr(),annot=True, cmap="coolwarm")
plt.title("Correlation graph")
plt.show()

# Distribution of Age
plt.figure(figsize=(10,5))
snb.histplot(df['Age'], bins=20, kde=True)
plt.title("Distribution of Age")
plt.show()

# Distribution of Frequent Cold
plt.figure(figsize=(10,5))
snb.countplot(x=df['Frequent Cold'])
plt.title("Distribution of Frequent Cold")
plt.show()

# Boxplot for Obesity
plt.figure(figsize=(8,5))
snb.boxplot(x=df['Obesity'])
plt.title("Outliers in Obesity ")
plt.show()
```
